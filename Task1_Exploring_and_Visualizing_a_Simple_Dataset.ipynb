# Explore the Iris dataset with pandas, seaborn, and matplotlib.


# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style and palette
sns.set_style("whitegrid")
sns.set_palette("Set2")
plt.rcParams['figure.dpi'] = 100
```

# Load Dataset
# Load Iris dataset (150 samples, 4 features, 3 species).

```python
# Load dataset
iris = sns.load_dataset('iris')
print("âœ… Dataset loaded!")
```

# Dataset Overview
# Inspect structure, missing values, and summary stats.

```python
# Dataset structure
print("Shape:", iris.shape)
print("Columns:", iris.columns.tolist())
print("First 5 rows:")
display(iris.head())
```

```python
# Data quality check
print("Missing Values:\n", iris.isnull().sum())
print("\nSummary Stats:")
display(iris.describe())
print("\nSpecies Count:")
display(iris['species'].value_counts())
```

# Correlation Heatmap
# Visualize feature correlations.

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(iris.iloc[:, :-1].corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

# Scatter Plot
# Plot sepal length vs. width by species.

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', size='species', sizes=(50, 150))
plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

# Histograms
# Show feature distributions.

```python
# Histograms
plt.figure(figsize=(10, 8))
for i, col in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=iris, x=col, kde=True, bins=20)
    plt.title(f'{col.replace("_", " ").title()}')
    plt.xlabel(f'{col.replace("_", " ").title()} (cm)')
plt.tight_layout()
plt.show()
```

# Box Plots
# Display feature spread and outliers by species.

```python
# Box plots
plt.figure(figsize=(10, 6))
iris_melted = pd.melt(iris, id_vars='species', value_vars=iris.columns[:-1])
sns.boxplot(data=iris_melted, x='variable', y='value', hue='species')
plt.title('Feature Distribution by Species')
plt.xlabel('Features')
plt.ylabel('Values (cm)')
plt.xticks(rotation=45)
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

# Pair Plot
# Visualize all feature pairs.


```python
# Pair plot
sns.pairplot(iris, hue='species', diag_kind='kde', height=2.5)
plt.suptitle('Pairwise Feature Relationships', y=1.02)
plt.show()
```

# Key Insights
# - **Data Quality**: No missing values, balanced species.
# - **Correlations**: Strong petal length-width correlation.
# - **Patterns**: Setosa distinct; versicolor/virginica overlap in sepals.
# - **Outliers**: Minor outliers in sepal width for virginica.
